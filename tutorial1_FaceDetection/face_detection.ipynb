تصدير الحزم الضرورية 

```python
import cv2
from matplotlib.pyplot import imshow
%matplotlib inline
```

سحب الملفات الضرورية التي تحتوي على خصائص الوجه

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

تحميل الصورة في متغير حيث يمكن وضع أي صورة 

```python
image = cv2.imread('./img.jpg')
imshow(image, cmap='gray')
```

هنا نقوم بتحويل الصورة إلى تدرج رمادي لسهولة التعامل مع الصور الرمادية

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

imshow(gray, cmap='gray')
```


هنا نقوم بتطبيق العملية وتخزين نقاط احداثيات الوجه على المستوى الإحداثي

```python
faces = face_cascade.detectMultiScale(gray, 1.3, 5)
```

في حال عدم العثور على وجه

```python
if faces is ():
    print("لا يوجد")
```

 بعد تخزين نقاط الإحداثيات، نقوم بتوصيل النقاط للرسم

```python

for (x,y,w,h) in faces:
    cv2.rectangle(image, (x,y), (x+w,y+h), (127,0,255), 2)
    imshow(image)
```
